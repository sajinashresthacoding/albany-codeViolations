```python
import geopandas as gpd
import pandas as pd

# --- STEP 1: Load neighborhood polygons ---
neighborhoods = gpd.read_file("neighborhood_boundaires.geojsonl.json")  # your polygon file
print(neighborhoods.columns)  # check for the column with neighborhood names
# let's assume the column is 'name'; replace with actual column if different

# --- STEP 2: Load addresses CSV ---
addresses = pd.read_csv("addresslayer.csv")  # must have Latitude and Longitude columns
print(addresses.head())

# --- STEP 3: Convert addresses to GeoDataFrame ---
address_gdf = gpd.GeoDataFrame(
    addresses,
    geometry=gpd.points_from_xy(addresses["LON"], addresses["LAT"]),
    crs="EPSG:4326"  # WGS84
)

# --- STEP 4: Ensure both datasets use the same CRS ---
neighborhoods = neighborhoods.to_crs(address_gdf.crs)

# --- STEP 5: Spatial join (assign each point to the polygon it falls in) ---
joined = gpd.sjoin(address_gdf, neighborhoods, how="left", predicate="within")

# Check the joined data
print(joined.head())

# --- STEP 6: Count points per neighborhood ---
# Replace 'name' with the actual column name of the neighborhood
count_by_neighborhood = (
    joined.groupby("Name")  # replace 'name' with your column
    .size()
    .reset_index(name="address_count")
)

print(count_by_neighborhood)

# --- STEP 7: Optional - save results ---
count_by_neighborhood.to_csv("address_counts_by_neighborhood.csv", index=False)
print("✅ Saved counts to address_counts_by_neighborhood.csv")
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# STEP 1: Load neighborhood polygons
# -----------------------------
neighborhoods = gpd.read_file("neighborhood_boundaires.geojsonl.json")
print("Neighborhood columns:", neighborhoods.columns)

# Replace with the actual column name of neighborhood names
neighborhood_name_col = "Name"  # <-- change this if needed

# -----------------------------
# STEP 2: Load addresses CSV
# -----------------------------
addresses = pd.read_csv("addresslayer.csv")
print(addresses.head())

# Replace with your lat/lon column names if different
lat_col = "LAT"
lon_col = "LON"

# -----------------------------
# STEP 3: Convert addresses to GeoDataFrame
# -----------------------------
address_gdf = gpd.GeoDataFrame(
    addresses,
    geometry=gpd.points_from_xy(addresses[lon_col], addresses[lat_col]),
    crs="EPSG:4326"  # assume lat/lon
)

# -----------------------------
# STEP 4: Ensure CRS match
# -----------------------------
neighborhoods = neighborhoods.to_crs(address_gdf.crs)

# -----------------------------
# STEP 5: Spatial join (points in polygons)
# -----------------------------
# Use 'intersects' to catch points on edges
joined = gpd.sjoin(address_gdf, neighborhoods, how="left", predicate="intersects")

# -----------------------------
# STEP 6: Count points per neighborhood
# -----------------------------
count_by_neighborhood = (
    joined.groupby(neighborhood_name_col)
    .size()
    .reset_index(name="address_count")
)

print("Counts by neighborhood:\n", count_by_neighborhood)

# -----------------------------
# STEP 7: Optional - Flag points outside all polygons
# -----------------------------
outside_points = joined[joined[neighborhood_name_col].isna()]
print(f"Points outside all neighborhoods: {len(outside_points)}")

# -----------------------------
# STEP 8: Save results
# -----------------------------
count_by_neighborhood.to_csv("address_counts_by_neighborhood.csv", index=False)
outside_points.to_csv("addresses_outside_neighborhoods.csv", index=False)

# -----------------------------
# STEP 9: Optional - Plot map
# -----------------------------
neighborhoods = neighborhoods.merge(count_by_neighborhood, on=neighborhood_name_col, how='left')
neighborhoods.plot(column='address_count', cmap='OrRd', legend=True, edgecolor='black')
address_gdf.plot(marker='o', color='blue', markersize=5, alpha=0.6)
plt.title("Addresses by Neighborhood")
plt.show()
```

```python
import geopandas as gpd
import pandas as pd

# -------------------------------
# 1. Load polygons and points
# -------------------------------
neigh = gpd.read_file("Untitled layer.geojson")  # Your neighborhoods GeoJSON
points = gpd.read_file("addresses.csv.geojson")  # Your points GeoJSON

# -------------------------------
# 2. Ensure CRS matches and is WGS84
# -------------------------------
neigh = neigh.to_crs("EPSG:4326")
points = points.to_crs("EPSG:4326")

# -------------------------------
# 3. Add unique ID to polygons
# -------------------------------
neigh['id'] = neigh.index.astype(str)

# -------------------------------
# 4. Spatial join: assign neighborhood ID to each point
# -------------------------------
points_with_neigh = gpd.sjoin(
    points, 
    neigh[['id', 'Name', 'geometry']], 
    how='left', 
    predicate='within'
)

# Rename columns for clarity
points_with_neigh = points_with_neigh.rename(columns={'Name_right': 'NeighborhoodName'})

# -------------------------------
# 5. Export points CSV for Flourish
# -------------------------------
points_with_neigh[['MainAddress', 'LAT', 'LON', 'id', 'NeighborhoodName']].to_csv(
    "addresses_points.csv", index=False
)

# -------------------------------
# 6. Create Regions table for Flourish
# -------------------------------
regions_table = neigh[['id', 'Name']].rename(columns={'Name': 'NeighborhoodName'})
regions_table.to_csv("regions.csv", index=False)

# -------------------------------
# 7. Save polygons with ID for Flourish
# -------------------------------
neigh.to_file("neighborhoods_with_id.geojson", driver="GeoJSON")

print("✅ Finished! Files ready for Flourish:")
print(" - addresses_points.csv")
print(" - regions.csv")
print(" - neighborhoods_with_id.geojson")
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1. Load neighborhoods polygons
# -------------------------------
neigh = gpd.read_file("Untitled layer.geojson")  # Replace with your file name

# Ensure CRS is WGS84
neigh = neigh.to_crs("EPSG:4326")

# Fix invalid geometries if any
neigh['geometry'] = neigh['geometry'].buffer(0)

# Add unique ID to each neighborhood
neigh['id'] = neigh.index.astype(str)

# Ensure property column for Flourish
if 'Name' not in neigh.columns:
    neigh['NeighborhoodName'] = neigh['id']
else:
    neigh['NeighborhoodName'] = neigh['Name']

# Keep only necessary columns for GeoJSON
neigh_export = neigh[['id', 'NeighborhoodName', 'geometry']]

# Export cleaned neighborhoods GeoJSON for Flourish
neigh_export.to_file("neighborhoods_with_id.geojson", driver="GeoJSON")
print("✅ Saved neighborhoods_with_id.geojson")

# -------------------------------
# 2. Load points (addresses)
# -------------------------------
points = gpd.read_file("addresses.csv.geojson")  # Replace with your file name

# Ensure same CRS
points = points.to_crs("EPSG:4326")

# -------------------------------
# 3. Spatial join: assign neighborhood ID to points
# -------------------------------
points_with_neigh = gpd.sjoin(
    points, 
    neigh_export[['id', 'NeighborhoodName', 'geometry']], 
    how='left', 
    predicate='within'
)

# Rename columns for clarity
points_with_neigh = points_with_neigh.rename(columns={'Name_left': 'OriginalName', 'Name_right': 'NeighborhoodName'})

# Ensure LAT and LON columns exist (Flourish requires them)
if 'LAT' not in points_with_neigh.columns or 'LON' not in points_with_neigh.columns:
    # Extract coordinates from geometry
    points_with_neigh['LON'] = points_with_neigh.geometry.x
    points_with_neigh['LAT'] = points_with_neigh.geometry.y

# Export points CSV for Flourish
points_with_neigh[['MainAddress', 'LAT', 'LON', 'id', 'NeighborhoodName']].to_csv(
    "addresses_points.csv", index=False
)
print("✅ Saved addresses_points.csv")

# -------------------------------
# 4. Create Regions table CSV for Flourish
# -------------------------------
regions_table = neigh_export[['id', 'NeighborhoodName']]
regions_table.to_csv("regions.csv", index=False)
print("✅ Saved regions.csv")

# -------------------------------
# 5. Optional: quick plot to check
# -------------------------------
plt.figure(figsize=(10,10))
neigh_export.plot(edgecolor='black', facecolor='none')
points_with_neigh.plot(x='LON', y='LAT', color='red', markersize=10)
plt.title("Preview: Neighborhoods and Points")
plt.show()

print("✅ All files ready for Flourish!")
```
