```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# --- Step 1: Load CSV points ---
df = pd.read_csv("data/2024 data for mapping.csv")

# Auto-detect lat/lon columns (assumes one column contains 'lat' and one contains 'lon')
lat_col = [c for c in df.columns if 'lat' in c.lower()][0]
lon_col = [c for c in df.columns if 'lon' in c.lower()][0]

# Convert to GeoDataFrame
geometry = [Point(xy) for xy in zip(df[lon_col], df[lat_col])]
gdf_points = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

# --- Step 2: Load GeoJSON polygons ---
gdf_polygons = gpd.read_file("data/gW3PK--insert-title-here-.geojson")

# Create a unique polygon ID
gdf_polygons['polygon_id'] = range(len(gdf_polygons))

# Ensure same CRS
gdf_points = gdf_points.to_crs(gdf_polygons.crs)

# --- Step 3: Spatial join (points inside polygons) ---
points_in_polygons = gpd.sjoin(gdf_points, gdf_polygons[['polygon_id', 'geometry']], 
                               how="inner", predicate='within')

# --- Step 4: Export CSV for Flourish ---
# Include polygon_id so you can group/color points in Flourish
points_in_polygons.to_csv("points_in_polygons_for_flourish.csv", index=False)

# --- Step 5: Export GeoJSON polygons with point counts ---
points_count = points_in_polygons.groupby('polygon_id').size().reset_index(name='point_count')
combined = gdf_polygons.merge(points_count, on='polygon_id', how='left')
combined['point_count'] = combined['point_count'].fillna(0).astype(int)
combined.to_file("polygons_with_points.geojson", driver="GeoJSON")

print(" Export complete:")
print(" - CSV for Flourish: points_in_polygons_for_flourish.csv")
print(" - GeoJSON polygons with counts: polygons_with_points.geojson")
```
